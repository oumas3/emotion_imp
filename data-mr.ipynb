```python
import pandas as pd

# Correct TweetEval sentiment paths
sentiment_splits = {
    "train": "sentiment/train-00000-of-00001.parquet",
    "test": "sentiment/test-00000-of-00001.parquet",
    "validation": "sentiment/validation-00000-of-00001.parquet"
}

# Load train only (later you can merge others)
tweeteval_df = pd.read_parquet("hf://datasets/cardiffnlp/tweet_eval/" + sentiment_splits["train"])

# Rename columns to unified format
tweeteval_df = tweeteval_df.rename(columns={
    "text": "text",
    "label": "label"
})

# Add source
tweeteval_df["source"] = "tweet_eval"
```

```python
fifa_df = pd.read_csv("/kaggle/input/fifa-world-cup-2022-tweets/fifa_world_cup_2022_tweets.csv")

# Rename to match our structure
fifa_df = fifa_df.rename(columns={
    "Tweet": "text",
    "Sentiment": "label"
})

# Keep only needed columns
fifa_df = fifa_df[["text", "label"]]

# Add source
fifa_df["source"] = "fifa"
```

```python
sentiment_map = {
    "Negative": 0,
    "Neutral": 1,
    "Positive": 2
}

fifa_df["label"] = fifa_df["label"].map(sentiment_map)

```

```python
for col in ["event_cluster", "emotion", "impact_weight"]:
    tweeteval_df[col] = None
    fifa_df[col] = None
```

```python
combined_df = pd.concat([tweeteval_df, fifa_df], ignore_index=True)

print("Merged dataset shape:", combined_df.shape)
combined_df.head()
```
